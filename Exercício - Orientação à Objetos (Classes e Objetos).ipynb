```python
class Veiculo:
    def __init__(self, tipo, modelo, fabricante, capacidade,
               tipoCombustivel, quantidadePortas, velocidade):
        self.tipo = tipo
        self.modelo = modelo
        self.fabricante = fabricante
        self.capacidade = capacidade
        self.tipoCombustivel = tipoCombustivel
        self.qtdPortas = quantidadePortas
        self.velMax = velocidade
        self.velAtual = 0
        self.combustivel = 55
        self.marcha = 3
    
    def acelerar(self, km):
        combustivelInicial = self.combustivel
        combustivelGasto = km/10
        combustivelFinal = self.combustivel - combustivelGasto
        self.combustivel = combustivelFinal
        print(f'O carro andou {km}Km e gastou {combustivelGasto}L de gasolina. \n\
Antes o tanque tinha {combustivelInicial}L, agora tem {combustivelFinal}.\n')
        
    def abastecer(self, litros):
        combustivelInicial = self.combustivel
        combustivelFinal = combustivelInicial + litros
        self.combustivel = combustivelFinal
        print(f'Você abasteceu {litros}L. \n\
Antes o tanque tinha {combustivelInicial}L de gasolina, agora tem {combustivelFinal}L.\n')
        
    def freiar(self, vel):
        self.velAtual = 50
        velInicial = self.velAtual
        velFinal = self.velAtual - vel
        self.velAtual = velFinal
        print(f'Você freiou e o carro foi de {velInicial}Km/h para {velFinal}Km/h.\n')

    def mudarMarcha(self, novaMarcha):
        if novaMarcha == 'n':
            novaMarcha = 0
            if self.marcha == 'n':
                print('O carro já está em ponto morto.')
            else:
                for marchaAtual in range(self.marcha, novaMarcha-1, -1):
                    if marchaAtual == 0:
                        print('Carro atualmente em ponto morto.')
                    else:
                        print(f'Carro atualmente na marcha {marchaAtual}.')
                    

if __name__ == '__main__':
    objVeiculo = Veiculo('Carro', 'Uno Mille', 'Fiat', 5, 'Gasolina', 4, 140)
    print(f'Tipo: {objVeiculo.tipo} \nModelo: {objVeiculo.modelo} \n\
Fabricante: {objVeiculo.fabricante} \nCapacidade: {objVeiculo.capacidade} pessoas \n\
Tipo do Combustível: {objVeiculo.tipoCombustivel} \nQuantidade de Portas: {objVeiculo.qtdPortas} portas\n\
Velocidade Máxima: {objVeiculo.velMax}Km/h\n')
    objVeiculo.acelerar(50)
    objVeiculo.abastecer(10)
    objVeiculo.freiar(25)
    objVeiculo.mudarMarcha('n')
```

```python
class Elevador:
    def __init__(self, andares, capacidade, pessoasPresentes):
        self.terreo = 0
        self.andares = andares
        self.capacidade = capacidade
        self.pessoasPresentes = pessoasPresentes

    def inicializar(self):
        print(f'Capacidade do elevador: {self.capacidade} pessoas \n\
Total de andares: {self.andares}\n')
        
    def entrar(self):
        if self.pessoasPresentes < self.capacidade:
            self.pessoasPresentes += 1
            print(f'Uma pessoa entrou no elevador \n\
Total de pessoas atualmente: {self.pessoasPresentes}\n')
        else:
            print('Elevedor lotado\n')

    def sair(self):
        if self.pessoasPresentes > 0:
            self.pessoasPresentes -= 1
            print(f'Uma pessoa saiu do elevador \n\
Total de pessoas atualmente: {self.pessoasPresentes}\n')
        else:
            print('Elevedor vazio\n')

    def subir(self, andarFinal, andarAtual=10):
        if andarFinal < andarAtual:
            print('Só é possível subir pra um andar acima do atual')
        elif andarAtual < self.andares:
            while andarAtual <= andarFinal:
                print(f'Elevador atualmente no andar {andarAtual}')
                andarAtual += 1
        else:
            print('Elevador já está no andar máximo')

    def descer(self, andarFinal, andarAtual=0):
        if andarFinal > andarAtual:
            print('Só é possível descer pra um andar abaixo do atual')
        elif(andarAtual > self.terreo):
            while andarAtual >= andarFinal:
                print(f'Elevador atualmente no andar {andarAtual}')
                andarAtual -= 1
        else:
            print('Elevador já está no térreo')

if __name__ == '__main__':
    objElevador = Elevador(10, 8, 0)
    objElevador.inicializar()
    objElevador.entrar()
    objElevador.sair()
    objElevador.subir(10)
    objElevador.descer(0)
    
```
