## Criando um dataframe usando matrizes

* Matriz 2x3

```python
import numpy as np

matriz = np.array([[1,2,3],[4,5,6]], np.int32)
df = pd.DataFrame(matriz, columns=['coluna1', 'coluna2', 'coluna3'])
```

```python
df
```

* Exibindo apenas os dados da coluna 3 onde o valor da coluna 2 é igual a 5 utilizando o método *loc*

```python
exibirColuna = df.loc[df['coluna2'] == 5, 'coluna3']
print(exibirColuna)
```

## Criando um dataframe usando Series de objetos

```python
nomesPaises = pd.Series(['China', 'India', 'Brasil', 'Estados Unidos',
                         'Japão', 'Reino Unido', 'Argentina'])
populacaoPaises = pd.Series([1412000000, 1408000000, 21400000, 331900000,
                             127500000, 67330000, 45810000])

paises = pd.DataFrame({'Nome dos paises':nomesPaises, 
                       'Populacao':populacaoPaises})
```

```python
paises
```

## Manipulando colunas e linhas de um dataframe

```python
dadosPessoais = {'Nome': ['Paula', 'José', 'Maria', 'João', 'Pedro', 'Ana'],
                 'Idade': [20, 23, 54, 36, 26, 57],
                 'Cidade': ['Aracaju', 'Salvador', 'Recife', 'Curitiba', 
                            'São Paulo', 'Rio de Janeiro'], 
                 'Nota': [8.3, 2.5, 8.7, 8.1, 5.2, 5.3]}

df = pd.DataFrame(dadosPessoais)
df
```

### Selecionando colunas 

* Forma 1

```python
print(df['Nome'])
```

* Forma 2

```python
print(df.Cidade)
```

* Múltiplas colunas

```python
print(df[['Nome', 'Cidade']])
```

### Iterando sobre colunas criando uma lista de colunas

```python
colunas = list(df) 
for i in colunas:
  print(df[i][2])
```

### Selecionando linhas

```python
linhasCidade = df['Cidade']
print(linhasCidade)
```

### Iterando sobre linhas utilizando o método iterrows()

```python
for i, j in df.iterrows():
  print(i,j)
  print()
```

## Criando um dataframe a partir da leitura do arquivo **nba.csv**

```python
dados = pd.read_csv('/content/nba.csv')
dados
```

```python
dadosArquivo = pd.read_csv('/content/nba.csv', index_col='Name')
```

```python
dadosArquivo
```

* Recuperando os dados utilizando slice ([ ])

```python
dadosArquivo[:2] 
```

* Recuperando os nomes e as idades dos 3 primeiros

```python
dadosArquivo[:3]['Age'] 
```

```python
dadosArquivo[:0]['Avery Bradley'] 
```

* Recuperando as linhas utilizando o método *.loc*, isto é, rótulo ou declaração condicional 


```python
primeiraLinha = dadosArquivo.loc['Avery Bradley']
segundaLinha = dadosArquivo.loc['Jae Crowder']
terceiraLinha = dadosArquivo.loc['John Holland']

print(primeiraLinha, '\n\n\n', segundaLinha, '\n\n\n', terceiraLinha)
```

* Recuperando uma única linha utilizando o método *.iloc*, isto é, números de linha 

```python
linha1 = dadosArquivo.iloc[0]
print(linha1)
```

* Recuperando os nomes e as idades com menos de 25 anos





```python
dadosArquivo.loc[dadosArquivo['Age'] < 25.0, ['Age']]
```
