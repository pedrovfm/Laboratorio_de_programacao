# **Orientação a Objetos**

* Paradigma de programação baseado no conceito de classes e objetos
  * Classes contém dados/atributos/propriedades na forma de campos e código/métodos na forma de procedimentos
  * Os programas são projetados a partir de objetos que interagem uns com os outros quando são manipulados

* Modelo de programação que reflete melhor o mundo real
  * Descrevem de maneira mais precisa os dados
  * Mais fácil de compreender e modelar o problema no código, desde a arquitetura até a implementação

* Facilita a reutilização de código
* Pequenas mudanças nos requisitos não implicam em grandes alterações no sistema em desenvolvimento

▶ **MÉTODOS** :

* Representam os comportamentos de uma Classe
* Permitem acesso aos atributos, tanto para recuperar os valores, como para alterá-los caso necessário
* Podem retornar ou não algum valor
* Podem possuir ou não parâmetros


````python
class Pessoa:
  def __init__(self, nome, idade, peso):
    self.nome = nome
    self.idade = idade
    self.peso = peso

  def dizerOla(self):
    print(f'Ola, meu nome é {self.nome}')

  def cozinhar(self, receita:str):
    print(f'Estou cozinhando {receita}')

  def correr(self, distancia:float):
    print(f'Corri {distancia} metros')

if __name__ == '__main__':
  objetoPessoa = Pessoa('João', 45, 82)
  objetoPessoa.dizerOla()
  objetoPessoa.cozinhar('Spaghetti')
  objetoPessoa.correr(80.0)

  print(objetoPessoa.nome)
  print(objetoPessoa.idade)
  print(objetoPessoa.peso)
````

### CARRO

* Classe
  * Atributos
    * cor, tipoCombustivel, potencia, quantidadePortas
 * Construtor
 * Métodos: 
    * ligar()
    * desligar()
    * acelerar()
    * freiar()
  * Instanciar um objeto, depois chamar todos os atributos e métodos, ou seja, if __name__ == '__main__':

````python
class Carro:
  def __init__(self, cor, tipoCombustivel, potencia, quantidadePortas):
    self.cor = cor
    self.tipoCombustivel = tipoCombustivel
    self.potencia = potencia
    self.quantidadePortas = quantidadePortas
    self.ligado = False
    self.velocidade = 0

  def ligar(self):
    if not self.ligado:
      self.ligado = True
      print('O carro esta ligado')
    else:
      print('O carro esta desligado')

  def desligar(self):
    if self.ligado:
      self.ligado = False
      self.velocidade = 0
      print('O carro foi desligado')
    else:
      print('O carro ja estava desligado')

  def acelerar(self, novaVelocidade:int):
    if self.ligado == True:
      self.velocidade += novaVelocidade
      print(f'O carro acelerou para {self.velocidade}')
    else:
      print('Nao e possivel acelerar')
      
  def freiar(self, novaVelocidade:int):
    if self.ligado:
      if self.velocidade >= novaVelocidade:
        self.velocidade -= novaVelocidade
      else:
        self.velocidade = 0
      print(f'O carro freiou para {self.velocidade}')
    else:
      print('Nao e possivel freiar')
        

if __name__ == '__main__':
   meuCarro = Carro('vermelho', 'gasolina', 150, 4)
   
   meuCarro.ligar()
   meuCarro.acelerar(50)
   meuCarro.acelerar(30)
   meuCarro.freiar(20)
   meuCarro.desligar()
````

### CONTA

* Classe
  * Atributos: numero
  * Construtor
  * Métodos: 
    * debitar(valor:int)
    * creditar(valor:int)
  * Instanciar um objeto, depois chamar todos os atributos e métodos, ou seja, if __name__ == '__main__':

````python
class Conta:

  def __init__(self, numero):
    self.numero = numero
    self.saldo = 0.0
  
  def debitar(self, valor):
    self.saldo -= valor

  def creditar(self, valor):
    self.saldo += valor

if __name__ == '__main__':
  conta = Conta('1234')
  print(conta.saldo)

  conta.creditar(100)
  print(conta.saldo)

  conta.debitar(50)
  print(conta.saldo)
````
